```python
import networkx as nx
from networkx.generators.community import LFR_benchmark_graph
import networkx.algorithms.community as nxcomm
import math
```

```python
n = 5000
tau1 = 3
tau2 = 1.5
mu = 0.2
G = LFR_benchmark_graph(
    n, tau1, tau2, mu, average_degree=7, min_community=30, seed=10
)
```

```python
communities = {frozenset(G.nodes[v]["community"]) for v in G}
```

```python
def to_communities(part):
    communities = []
    for i in range(max(part.values()) + 1):
    #for i in range(3):
        communities.append(frozenset(k for k,v in part.items() if v == i))
    return communities
    
```

```python
print(len(communities))
nxcomm.modularity(G,communities)
#print(communities)
```

```python
import community as community_louvain
```

```python
partition = community_louvain.best_partition(G)
```

```python
#print(G)
communities_louvain = to_communities(partition)
print(len(communities_louvain))
```

```python
nxcomm.modularity(G,communities_louvain)
```

```python
res_limit_val = math.sqrt(2*G.number_of_edges())
print(res_limit_val)
for com in communities:
    sub = G.subgraph(com)
    if sub.number_of_edges() < res_limit_val / 2:
        print(sub.nodes)
```
